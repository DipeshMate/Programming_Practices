```python
data = {'col1': [1, 2, 3], 'col2': [4, 5, 6]}
index = ['a', 'b', 'c']
df = pd.DataFrame(data, index=index)
print(data)
    # Select row 'a' and all columns
print(df.loc['a'])
    
    # Select rows 'a' and 'b', and column 'col2'
print(df.loc[['a', 'b'], 'col2'])
    
    # Select rows from 'a' to 'c', and columns from 'col1' to 'col2'
print(df.loc['a':'c', 'col1':'col2'])
```
